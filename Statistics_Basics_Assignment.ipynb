5.Discuss the role of random sampling in making inferences about populations.

Ans - Random sampling plays a crucial role in statistics because it allows researchers to draw accurate and unbiased conclusions about a larger population using data from a smaller group (sample).

Key Roles of Random Sampling:

Ensures Representativeness
Each member of the population has an equal chance of being selected.

This helps the sample reflect the diversity and characteristics of the entire population.

Reduces Bias

Random sampling prevents selection bias, where certain groups might be over- or under-represented.

This leads to more reliable results.

Supports Valid Statistical Inference

Most statistical methods (e.g., confidence intervals, hypothesis tests) assume the sample is random.

This allows researchers to generalize findings from the sample to the entire population with known levels of accuracy.

Allows Use of Probability

Random sampling enables the application of probability theory, which is essential for measuring uncertainty in estimates (e.g., margins of error).

Example:

If a company randomly surveys 500 customers out of 10,000, the average satisfaction score from those 500 can be used to infer the likely satisfaction of all customersâ€”if the sample is randomly selected.



6.Explain the concept of skewness and its types. How does skewness affect the interpretation of data?


Ans - Skewness measures the asymmetry of a distribution around its mean. In a perfectly symmetrical distribution, the left and right sides mirror each other (like the bell-shaped normal distribution). Skewness tells us whether data is stretched more to one side.

Types of Skewness
1.Positive Skew (Right Skewed)
Tail is longer on the right

Mean > Median > Mode

Most values are concentrated on the left, with a few high outliers pulling the mean to the right.

Example: Income distribution (few very high earners)

2.Negative Skew (Left Skewed)
Tail is longer on the left

Mean < Median < Mode

Most values are concentrated on the right, with a few low outliers pulling the mean to the left.

Example: Age at retirement (most retire around the same age, but a few retire early)

3.Zero Skew (Symmetrical Distribution)
Tails are equal on both sides

Mean â‰ˆ Median â‰ˆ Mode

The data is evenly distributed.

Example: Standardized test scores in a large, balanced group

How Skewness Affects Data Interpretation
Distorts Measures of Central Tendency

In skewed data, the mean is pulled toward the tail.

Median is a better measure of central tendency in such cases.

Influences Statistical Assumptions

Many statistical tests (e.g., t-tests, ANOVA) assume normality (no skew).

Skewness may require data transformation or non-parametric methods.

Affects Visual Interpretation

Skewed data on a histogram or box plot can show concentration or outliers more clearly.

The normal distribution is a continuous probability distribution that is symmetric about the mean. Itâ€™s often referred to as the bell curve because of its shape. Here are the key properties:

Total Area = 1: The total area under the curve is equal to 1, representing the entire probability space.

The Empirical Rule applies specifically to normal distributions and describes how data is distributed in relation to the mean and standard deviations.

11.Explain what a random variable is and differentiate between discrete and continuous random variables.

Ans - A random variable is a numerical outcome of a random phenomenon or experiment. It represents a quantitative value that results from an uncertain event. A random variable can take different values based on the outcome of the random process.

For example, if you roll a fair die, the random variable could be the number that shows up on the die.

Types of Random Variables:

There are two main types of random variables:

1.Discrete Random Variables

Definition: A discrete random variable can take only a finite or countably infinite number of distinct values. These values are usually integers.

Key Characteristic: The possible values can be listed out, and there are gaps between them.

Example:

The number of heads in 5 coin flips (values: 0, 1, 2, 3, 4, 5).

The number of cars passing a toll booth in an hour (values: 0, 1, 2, 3, ...).

Probability Distribution: The probability mass function (PMF) gives the probability of each possible value.

2.Continuous Random Variables

Definition: A continuous random variable can take an infinite number of values within a given range. The values are not countable, and they can be measured rather than counted.

Key Characteristic: There are no gaps between the possible values, and the variable can take any value within an interval.

Example:

The height of a person (can be any value, like 5.4, 5.45, 5.445, etc.).

The time it takes to run a race (could be 12.34 seconds, 12.345, etc.).

Probability Distribution: The probability density function (PDF) is used to describe the probability distribution, but for continuous variables, the probability of any single exact value is zero (since there are infinite possible values).

Key Differences Between Discrete and Continuous Random Variables:

1.Nature of Values:

Discrete random variables take countable, distinct values. These values are often integers. For example, the number of students in a class or the number of heads when flipping a coin multiple times.

Continuous random variables, on the other hand, can take any value within a given range and are uncountable. These values are real numbers, and there are no gaps between them. Examples include height, weight, or time.

2.Examples:

Discrete: The number of books on a shelf (can be 1, 2, 3, etc.).

Continuous: The time it takes for a person to run a race (could be any value like 12.45 seconds, 12.456 seconds, etc.).

3.Probability Distribution:

For discrete random variables, the probability mass function (PMF) is used, which gives the probability of each possible value.

For continuous random variables, the probability density function (PDF) is used. Here, the probability of a specific value is technically zero, and probabilities are defined over ranges (intervals).

4.Probability of Exact Value:

Discrete random variables can have a non-zero probability for specific values. For example, you could calculate the probability of rolling a 3 on a die (1/6).

Continuous random variables have an infinitesimally small probability for any exact value. Since the range is infinite, the probability of any specific value is considered zero. Probability is instead calculated over intervals (e.g., the probability that a person's height is between 5'5" and 5'7").

5.Possible Outcomes:

Discrete variables have a finite or countably infinite number of possible outcomes. You can list all possible outcomes (e.g., the number of students can be 0, 1, 2, 3, and so on).

Continuous variables have an infinite number of outcomes within any given range, and these values can take on any value within that interval.

6.Measurement vs. Counting:

Discrete random variables are typically associated with counting (e.g., the number of cars in a parking lot).

Continuous random variables are typically associated with measurement (e.g., measuring the time it takes to finish a task or the weight of an object).
